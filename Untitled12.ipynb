```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("./Motor_Vehicle_Collisions_-_Crashes.csv")
```

**Univariate** **Analysis**

# **Histogram**


```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df["CRASH DATE"], bins=10)
plt.xlabel("date of crash")
plt.ylabel("Number of crashes")
plt.title("Histogram of car_crashes")
plt.show()
```

**Boxplot**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="NUMBER OF PERSONS KILLED", data=df)
plt.xlabel("Number_of_persons_killed")
plt.ylabel("Value")
plt.title("Boxplot of Number_of_people_killed")
plt.show()
```
